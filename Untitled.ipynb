```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

# Chargement des données
df = pd.read_csv('Data.csv', sep=";")  # Assurez-vous que le chemin est correct
# Filtrer les données pour l'entreprise 'CS.PA'
df_filtered = df[df['Symbole'] == 'CS.PA']

# Préparation des données
df_filtered['Date'] = pd.to_datetime(df_filtered['Date'], format='%Y-%m-%d')
df_filtered.sort_values('Date', inplace=True)

# Sélection des caractéristiques et de la cible
X = df_filtered[['Open', 'High', 'Low', 'Volume']]  # Variables indépendantes
y = df_filtered['Close']  # Variable à prédire

# Division en ensembles d'entraînement et de test
train_size = int(len(X) * 0.8)  # 80% pour l'entraînement, 20% pour le test
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Normalisation des données
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Régression Linéaire
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Génération des prédictions pour 60 jours
last_date = df_filtered['Date'].max()
future_dates = [last_date + timedelta(days=x) for x in range(1, 61)]
future_data = pd.DataFrame({
    'Open': np.repeat(X_train.mean()['Open'], 60),
    'High': np.repeat(X_train.mean()['High'], 60),
    'Low': np.repeat(X_train.mean()['Low'], 60),
    'Volume': np.repeat(X_train.mean()['Volume'], 60)
})

# Normalisation et prédiction des données futures
future_data_scaled = scaler.transform(future_data)
future_predictions = model.predict(future_data_scaled)

# Création du graphique
plt.figure(figsize=(12, 6))
plt.plot(df_filtered['Date'], df_filtered['Close'], label='Historical Close Prices', marker='o')
plt.plot(future_dates, future_predictions, label='Predicted Close Prices', marker='x', linestyle='--')
plt.title('Historical and Predicted Close Prices for CS.PA')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt

# Chargement des données
df = pd.read_csv('Data.csv', sep=";")  # Assurez-vous que le chemin est correct
# Filtrer les données pour l'entreprise 'CS.PA'
df_filtered = df[df['Symbole'] == 'CS.PA']

# Préparation des données
df_filtered['Date'] = pd.to_datetime(df_filtered['Date'], format='%Y-%m-%d')
df_filtered.sort_values('Date', inplace=True)

# Sélection des caractéristiques et de la cible
X = df_filtered[["Date"]]  # Variables indépendantes
y = df_filtered['Close']  # Variable à prédire

# Division des données en un ensemble d'entraînement et un ensemble de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Entraînement du modèle
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Prédiction sur l'ensemble de test
y_pred = regressor.predict(X_test)

# Évaluation du modèle
print('Erreur quadratique moyenne (RMSE):', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

# Création d'un DataFrame pour les jours futurs
future_days = pd.DataFrame(np.zeros((60, 4)), columns=X.columns)

# Prédiction pour les 60 prochains jours
future_pred = regressor.predict(future_days)

# Création du graphique
plt.figure(figsize=(10, 6))
plt.plot(df_filtered["Date"], df_filtered['Close'], label='Données historiques')
plt.plot(pd.date_range(start=df_filtered["Date"].max(), periods=60), future_pred, label='Prédictions pour les 60 prochains jours', color='red')
plt.xlabel('Jours')
plt.ylabel('Valeur prédite')
plt.title('Prédictions de régression linéaire pour les 60 prochains jours')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeRegressor  # Importer le modèle de régression par arbre de décision
from sklearn import metrics
import numpy as np
from pandas.tseries.offsets import BDay
from scipy import stats
import matplotlib.pyplot as plt

# Charger les données
df = pd.read_csv('fichiers/Data.csv',sep=";")
df = df[df['Symbole'] == 'KER.PA']
# Convertir la colonne de date en datetime et extraire le nombre de jours
df['Date'] = pd.to_datetime(df['Date'])
start_date = pd.to_datetime('2024-02-01')
df = df[df['Date'] >= start_date]
df['Date'] = (df['Date'] - df['Date'].min())  / np.timedelta64(1,'D')

# Définir X et y
X = df['Date'].values.reshape(-1,1)
y = df['Close'].values.reshape(-1,1)

# Diviser les données en ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Entraîner le modèle
regressor = DecisionTreeRegressor()  # Utiliser le modèle de régression par arbre de décision
regressor.fit(X_train, y_train)

# Le reste du code reste le même
# Définir la date de début pour les prévisions, en excluant les week-ends
start_forecast_date = pd.to_datetime('2024-03-11')

# Calculer le nombre de jours ouvrables à prédire
days_in_future = 30  # Le nombre total de jours à prédire
future_business_days = pd.date_range(start_forecast_date, periods=days_in_future, freq=BDay())

# Convertir les dates futures en nombres, basé sur le calcul original
future_forecast_days = [(day - (start_date + pd.Timedelta(days=df['Date'].min()))).days for day in future_business_days]
future_forecast = np.array(future_forecast_days).reshape(-1,1)

# Prédiction avec le modèle entraîné
y_pred = regressor.predict(future_forecast)

# Création d'un DataFrame pour les prédictions futures
future_predictions_df = pd.DataFrame({
    'Future Date': future_business_days,
    'Predicted Close': y_pred.flatten()
})

# Afficher les prédictions
print(future_predictions_df.head(10))



residuals = y_test - regressor.predict(X_test)

# Calculer l'erreur standard des résidus
stderr = residuals.std()

# Nombre de prédictions
n = future_forecast.shape[0]

# Degrés de liberté
df = len(X_test) - 2

# Trouver le t-score pour un intervalle de confiance de 95%
t_score = stats.t.ppf(0.975, df)

# Calculer l'intervalle de confiance pour chaque prédiction
confidence_interval = t_score * stderr / np.sqrt(n)

# Ajouter l'intervalle de confiance aux prédictions
future_predictions_df['Lower Bound'] = future_predictions_df['Predicted Close'] - confidence_interval
future_predictions_df['Upper Bound'] = future_predictions_df['Predicted Close'] + confidence_interval

# Afficher les prédictions avec l'intervalle de confiance
print(future_predictions_df.head(10))

# Sauvegarde du DataFrame dans un fichier CSV
future_predictions_path = 'future_predictions.csv'  # Adjusted path for saving in accessible location
future_predictions_df.to_csv(future_predictions_path, index=False)

lower_bound = future_predictions_df['Lower Bound']
upper_bound = future_predictions_df['Upper Bound']
# Création du graphique
plt.figure(figsize=(10, 6))
plt.plot(future_business_days, y_pred, color='blue', label='Prédictions linéaires')
plt.fill_between(future_business_days, lower_bound, upper_bound, color='gray', alpha=0.2, label='Intervalle de confiance 95%')
plt.xlabel('Date')
plt.ylabel('Prédiction Close')
plt.title('Prédictions avec intervalle de confiance à 95%')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, learning_curve
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
import matplotlib.pyplot as plt
import numpy as np
from pandas.tseries.offsets import BDay

# Fonction pour créer le modèle, nécessaire pour KerasRegressor
def create_model():
    model = Sequential()
    model.add(Dense(50, input_dim=1, activation='relu'))
    model.add(Dense(30, activation='relu'))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

# Charger les données
df = pd.read_csv('Data.csv',sep=";")
df = df[df['Symbole'] == 'KER.PA']
# Convertir la colonne de date en datetime et extraire le nombre de jours
df['Date'] = pd.to_datetime(df['Date'])
start_date = pd.to_datetime('2024-02-01')
df = df[df['Date'] >= start_date]
df['Date'] = (df['Date'] - df['Date'].min())  / np.timedelta64(1,'D')

# Définir X et y
X = df['Date'].values.reshape(-1,1)
y = df['Close'].values.reshape(-1,1)

# Diviser les données en ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Créer le modèle
model = KerasRegressor(build_fn=create_model, epochs=50, batch_size=10, verbose=0)

# Entraîner le modèle
model.fit(X_train, y_train)

# Fonction pour tracer les courbes d'apprentissage
def plot_learning_curve(model, X, y):
    train_sizes, train_scores, test_scores = learning_curve(model, X, y, cv=5)

    train_scores_mean = np.mean(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)

    plt.plot(train_sizes, train_scores_mean, label='Score d\'entraînement')
    plt.plot(train_sizes, test_scores_mean, label='Score de validation')

    plt.ylabel('Score')
    plt.xlabel('Taille de l\'échantillon d\'entraînement')
    plt.title('Courbes d\'apprentissage')
    plt.legend(loc='best')
    plt.show()

# Utilisez la fonction sur votre modèle
plot_learning_curve(model, X, y)
```
